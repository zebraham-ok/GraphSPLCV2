```python
from Neo4jHost import get_remote_driver
import pandas as pd
neo4j_host=get_remote_driver()
```

```python
result=neo4j_host.execute_query(query="""
        match (n:EntityObj)-[r:SupplyProductTo]->(m:EntityObj)
        where r.verified=true and r.analysing_process is not null and r.original_content is not null
        return n.name, n.industry_2nd, r.product, m.name, m.industry_2nd, r.analysing_process, r.original_content, r.url
        order by rand()
        limit 400
    """)
result_table=pd.DataFrame(result, columns=["supplier", "supplier_industry", "product", "customer", "customer_industry", "analysing_process", "original_content", "url"])
result_table
```

```python
result_table.to_excel("")
```

```python
result2=neo4j_host.execute_query(query="""
        match (n:EntityObj)-[r:SupplyProductTo]->(m:EntityObj)
        where r.analysing_process is not null and r.original_content is not null
        return n.name, n.industry_2nd, r.product, m.name, m.industry_2nd, r.analysing_process, r.original_content, r.url, r.verified
        order by rand()
        limit 800
    """)
result_table2=pd.DataFrame(result2, columns=["supplier", "supplier_industry", "product", "customer", "customer_industry", "analysing_process", "original_content", "url", "verified"])
result_table2
```

```python
(result_table2["verified"]==True).sum()
```

```python
import re

def clean_text(text):
    # 使用正则表达式移除非法字符
    return re.sub(r'[\x00-\x08\x0b-\x0c\x0e-\x1f]', '', text)

result_table2["original_content"]=result_table2["original_content"].apply(clean_text)
```
